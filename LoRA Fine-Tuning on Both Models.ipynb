```python
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM, BitsAndBytesConfig


model_names = {
    "llama3": "meta-llama/Meta-Llama-3-8B",
    "deepseek": "deepseek-ai/DeepSeek-R1-Distill-Llama-8B"
}

bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_compute_dtype=torch.float16,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_use_double_quant=True
)


tokenizers = {key: AutoTokenizer.from_pretrained(model) for key, model in model_names.items()}
models = {
    key: AutoModelForCausalLM.from_pretrained(
        model,
        quantization_config=bnb_config,
        device_map="auto"
    ) for key, model in model_names.items()
}

print(" Models & Tokenizers Loaded Successfully!")
```

```python
def generate_response(model, tokenizer, prompt, max_length=512):
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

    inputs = tokenizer(prompt, return_tensors="pt").to(device)

    with torch.no_grad():
        output = model.generate(
            **inputs,
            max_new_tokens=max_length,
            temperature=0.7,
            do_sample=True
        )

    return tokenizer.decode(output[0], skip_special_tokens=True)


prompt = "Analyze Apple's stock market performance based on recent news and financial data."


llama3_response = generate_response(models["llama3"], tokenizers["llama3"], prompt)
deepseek_response = generate_response(models["deepseek"], tokenizers["deepseek"], prompt)


print("\n📊 **Llama3-8B Analysis:**\n", llama3_response)
print("\n📊 **DeepSeek-R1 Analysis:**\n", deepseek_response)
```

```python
from peft import LoraConfig, get_peft_model
from datasets import load_dataset
from transformers import TrainingArguments, Trainer


lora_config = LoraConfig(
    r=8,
    lora_alpha=16,
    target_modules=["q_proj", "v_proj"],
    lora_dropout=0.1,
    bias="none",
    task_type="CAUSAL_LM"
)


lora_models = {key: get_peft_model(model, lora_config) for key, model in models.items()}


dataset = load_dataset("tatsu-lab/alpaca", split="train")


def tokenize_function(example):
    return tokenizers["llama3"](example["instruction"], padding="max_length", truncation=True)

tokenized_datasets = dataset.map(tokenize_function, batched=True)


training_args = TrainingArguments(
    output_dir="./results",
    per_device_train_batch_size=2,
    per_device_eval_batch_size=2,
    num_train_epochs=2,
    logging_steps=10,
    save_steps=100,
    save_total_limit=2,
    evaluation_strategy="steps",
    optim="adamw_torch"
)


trainers = {
    key: Trainer(
        model=lora_model,
        args=training_args,
        train_dataset=tokenized_datasets,
        eval_dataset=tokenized_datasets.select(range(100)),
        tokenizer=tokenizer
    ) for key, (lora_model, tokenizer) in zip(lora_models.keys(), zip(lora_models.values(), tokenizers.values()))
}

print(" LoRA Fine-Tuning Setup Completed!")
```

```python

print(" Fine-Tuning Llama3-8B...")
trainers["llama3"].train()

print("Fine-Tuning DeepSeek-R1...")
trainers["deepseek"].train()

```

```python

for key, trainer in trainers.items():
    trainer.save_model(f"./fine_tuned_{key}")

fine_tuned_models = {
    key: AutoModelForCausalLM.from_pretrained(f"./fine_tuned_{key}")
    for key in trainers.keys()
}

print(" Fine-Tuned Models Saved & Loaded Successfully!")
```
